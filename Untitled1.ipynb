```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("International_T20_Data.csv")
df.head()
```

```python
##BASIC OPERATIONS OF PANDAS
df.head()
df.tail()
df.sample(5)
df.shape
df.columns
df.index
df.dtypes
df.info()
df.describe()
```

```python
##DATA CLEANING
##-MISSING VALUES
df.isnull().sum()
# df.dropna(inplace=True) # Commenting out to prevent loss of all data

##Removing duplicates
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
##COLUMN OPERATIONS
df.rename(columns={'info.city':'City'}, inplace=True)

df['Runs_margin'] = df['info.outcome.by.runs']

df.sort_values(by='Runs_margin', ascending=False).head()
```

```python
##Filtering Data
df[df['info.gender'] == 'male']

df[df['info.outcome.by.runs'] > 50]

df.loc[:, ['City','info.venue']]
```

```python
##GroupBy Operations

##Matches per city
df.groupby('City').size()

##Average run margin by venue
df.groupby('info.venue')['info.outcome.by.runs'].mean()

##Matches won by each team
df['info.outcome.winner'].value_counts()
```

```python
##8️ NumPy Functions
runs = np.array(df['info.outcome.by.runs'])

np.mean(runs)
np.median(runs)
np.std(runs)
np.var(runs)
np.max(runs)
np.min(runs)
np.random.choice(runs,10)
```

```python
##Pivot Table
pd.pivot_table(df,
               values='info.outcome.by.runs',
               index='City',
               aggfunc=np.mean)
```

```python
##Correlation Analysis
df.corr(numeric_only=True)
```

```python
##Matplotlib Visualizations
teams = df['info.outcome.winner'].value_counts().head(10)

plt.figure(figsize=(10,6))
plt.bar(teams.index, teams.values)
plt.xticks(rotation=45)
plt.title("Top Teams by Wins")
plt.show()
```

```python
##HISTOGRAM
plt.hist(df['info.outcome.by.runs'], bins=20)
plt.title("Run Margin Distribution")
plt.xlabel("Runs")
plt.ylabel("Frequency")
plt.show()
```

```python
##pie chart
gender = df['info.gender'].value_counts()

plt.pie(gender.values, labels=gender.index, autopct='%1.1f%%')
plt.title("Match Distribution by Gender")
plt.show()
```

```python
##Seaborn Visualizations
##Count Plot
sns.countplot(x='info.toss.decision', data=df)
plt.title("Toss Decisions")
plt.show()
```

```python
##Box Plot
sns.boxplot(x='info.gender', y='info.outcome.by.runs', data=df)
plt.show()
```

```python
##Violin Plot
sns.violinplot(x='info.gender', y='info.outcome.by.runs', data=df)
plt.show()
```

```python
##Heatmap
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
